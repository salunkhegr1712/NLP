```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.geeksforgeeks.org/"
# requests.get will fetch the website and print the results

r=requests.get(url)
# r.content will provide the content inside the body of returned html code 
```

```python
# now as we get r and it has something content init 
htmlBeautified=BeautifulSoup(r.content,'html.parser')
print(htmlBeautified)
# so html5lib just do one thing here and that is to make html tree and parse the 
# html code and make it more modular and it is very much in view as compared to that of r.content file 
# which provide html in raw and unarranged manner 
```

```python
# get all links from all the anchor tags 

# say as you know each tag has own inbuilt values like a tag has values like 
from lib2to3.pytree import type_repr


l=htmlBeautified.find_all("a");
links_dataset=set()
for i in range(len(l)):
    # by using get method we can fetch through the inner attributes of the perticular tag like p a and many more like that 
    m=str(l[i].get("href"))
    if(m[0:5]=="https"):
        print(m)
        links_dataset.add(m)
    # links_dataset.append(l[i]["href"]);


links_dataset
```

```python
# lets see more of the types from the BeautifulSoup library and there parsed elments and all 
a="<p><!-- this is a comment--!><p>"
# lets see type of a we already know it by seeing it only and it is a string 
# <p> is paragraph tag and the <!-- --> is space where we can write html comments 
print(type(a))
```

```python
# lets convert the text into type of the BeautifulSoup 
soup=BeautifulSoup(a);
print(soup.p)
# lets print the comment inside of the p tag 
print(soup.p.string) #it is just like get_text() function 
# nameofSoup.its tag will give that perticular tag from that BeautifulSoup text 
# lets seee type of it now 
print(type(soup)) #<class 'bs4.BeautifulSoup'>
print(type(soup.p)) #<class 'bs4.element.Tag'>
print(type(soup.p.string))
```
