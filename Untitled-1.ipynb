```python
import requests
from bs4 import  BeautifulSoup
```

```python
url="https://indianexpress.com/"
# requests.get will fetch the website and print the results

r=requests.get(url)
# r.content will provide the content inside the body of returned html code 
r=r.content
```

```python
# now as we get r and it has something content init 
from distutils.command.install_egg_info import to_filename


htmlBeautified=BeautifulSoup(r,'html5lib')
# so html5lib just do one thing here and that is to make html tree and parse the 
# html code and make it more modular and it is very much in view as compared to that of r.content file 
# which provide html in raw and unarranged manner 
print(htmlBeautified)
```
